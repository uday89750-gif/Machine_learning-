```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
iris=pd.read_csv('iris.csv')
iris.sample(6)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}))
```

```python
plt.scatter(iris['SepalWidthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.7)
plt.xlabel('SepalWidthCm')
plt.ylabel('PetalLengthCm')
plt.title('Scatter Plot')
plt.show()
```

plot size

```python
plt.figure(figsize=(3,3))
plt.scatter(iris['SepalWidthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.7)
plt.xlabel('SepalWidthCm')
plt.ylabel('PetalLengthCm')
plt.title('Scatter Plot')
```

Annotation

```python
batter= pd.read_csv('batter.csv')
sample_df=batter.sample(25)

plt.scatter(sample_df['avg'],sample_df['strike_rate'])
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

h and v line

```python
batter= pd.read_csv('batter.csv')
sample_df=batter.sample(25)

plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.axhline(80,color='red',linestyle='--')
plt.axvline(10,color='red',linestyle='--')
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
fig ,ax = plt.subplots(nrows=2,ncols=1,figsize=(8,10),sharex=True);
ax[0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[0].set_xlabel('avg')
ax[0].set_ylabel('strike_rate')
ax[0].set_title('Scatter Plot')
ax[1].scatter(batter['avg'],batter['runs'])
ax[1].set_xlabel('avg')
ax[1].set_ylabel('strike_rate')
ax[1].set_title('Scatter Plot')
```
