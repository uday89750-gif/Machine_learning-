```python
import pandas as pd
import numpy as np
import sklearn as sk
```

```python
df= pd.read_csv("customer.csv",usecols=["education","review","purchased"])
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(df.drop("purchased",axis=1),df['purchased'],test_size=0.3)
x_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[['School','UG','PG'],['Poor','Average','Good']])

oe.fit(x_train[['education','review']])

x_train[['education','review']] = oe.transform(x_train[['education','review']])
x_test[['education','review']] = oe.transform(x_test[['education','review']])
```

```python
x_train = pd.DataFrame(x_train,columns=['review','education'])
x_test= pd.DataFrame(x_test,columns=['review','education'])
```

```python
#column transformer
df= pd.read_csv("covid_toy.csv")
df
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns=["has_covid"]),df["has_covid"],test_size=0.2)
```

```python
y_test
```

```python
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers =[( 'tnf1',SimpleImputer(),['fever']),
( 'tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
('tnf3',OneHotEncoder(sparse_output = False,drop='first'),['gender','city'])],remainder = 'passthrough')
```

```python
x_train = transformer.fit_transform(x_train)
x_test = transformer.transform(x_test)
```

```python
x_train = pd.DataFrame(x_train,columns = transformer.get_feature_names_out())
x_test = pd.DataFrame(x_test,columns = transformer.get_feature_names_out())
```
